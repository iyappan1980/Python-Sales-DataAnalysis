```python
#  Derived Metrics
df['tip_pct'] = df['tip'] / df['total_bill'] * 100
print(df[['total_bill', 'tip', 'tip_pct']])
```

```python
df
```

```python
# Remove duplicate rows
df_cleaned = df.drop_duplicates()
```

```python
# Compute correlation matrix
corr_matrix = df_cleaned.corr()

# Plot heatmap
plt.figure(figsize=(6, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
### 5 Useful Pivot Tables ###

# 1. Average Tip Percentage by Day
pivot1 = df_cleaned.pivot_table(values='tip_percentage', index='day', aggfunc='mean')
print("\n1. Average Tip Percentage by Day:\n")
pivot1
```
