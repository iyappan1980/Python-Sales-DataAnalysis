```python
# Check for duplicate rows
duplicates = df.duplicated()
```

```python
duplicates
```

```python
# Print duplicate rows, if any
if duplicates.any():
    print("Duplicate Rows Found:\n", df[duplicates])
else:
    print("No duplicate rows found.")
```

```python
# Remove duplicate rows
df_cleaned = df.drop_duplicates()
```

```python
df
```

```python
df_cleaned
```

```python

# Feature Engineering: Adding new columns
df_cleaned['tip_percentage'] = (df_cleaned['tip'] / df_cleaned['total_bill']) * 100  # Tip percentage
df_cleaned['total_per_person'] = df_cleaned['total_bill'] / df_cleaned['size']  # Bill per person
df_cleaned['is_weekend'] = df_cleaned['day'].apply(lambda x: 1 if x in [2, 3] else 0)  # Weekend flag (1 for Sat/Sun)
```

```python
df_cleaned
```

```python
# Compute correlation matrix
corr_matrix = df_cleaned.corr()

# Plot heatmap
plt.figure(figsize=(6, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
### 5 Useful Pivot Tables ###

# 1. Average Tip Percentage by Day
pivot1 = df_cleaned.pivot_table(values='tip_percentage', index='day', aggfunc='mean')
print("\n1. Average Tip Percentage by Day:\n")
pivot1
```

```python
# 2. Total Bill Amount by Day and Size
pivot2 = df_cleaned.pivot_table(values='total_bill', index='day', columns='size', aggfunc='sum', fill_value=0)
print("\n2. Total Bill Amount by Day and Size:\n", )
pivot2
```

```python



# 3. Average Tip Amount by Sex and Smoker Status
pivot3 = df_cleaned.pivot_table(values='tip', index='sex', columns='smoker', aggfunc='mean', fill_value=0)
print("\n3. Average Tip Amount by Sex and Smoker Status:\n", pivot3)
```

```python

# 4. Total Bill and Tip Percentage by Weekend or Weekday
pivot4 = df_cleaned.pivot_table(values=['total_bill', 'tip_percentage'], index='is_weekend', aggfunc='mean')
print("\n4. Total Bill and Tip Percentage by Weekend or Weekday:\n", pivot4)
```

```python
# 5. Number of Customers Served by Day
pivot5 = df_cleaned.pivot_table(values='size', index='day', aggfunc='sum')
print("\n5. Number of Customers Served by Day:\n", pivot5)
```

```python
# Plot 1: Average Tip Percentage by Day
plt.figure(figsize=(10, 5))
sns.barplot(x=pivot1.index, y=pivot1['tip_percentage'], palette='coolwarm')
plt.xlabel('Day')
plt.ylabel('Average Tip Percentage')
plt.title('Average Tip Percentage by Day')
plt.show()

# Plot 2: Total Bill Amount by Day and Size
plt.figure(figsize=(10, 5))
sns.heatmap(pivot2, annot=True, cmap='Blues', fmt=".2f", linewidths=0.5)
plt.xlabel('Table Size')
plt.ylabel('Day')
plt.title('Total Bill Amount by Day and Table Size')
plt.show()

# Plot 3: Average Tip Amount by Sex and Smoker Status
plt.figure(figsize=(8, 5))
sns.heatmap(pivot3, annot=True, cmap='OrRd', fmt=".2f", linewidths=0.5)
plt.xlabel('Smoker (0=No, 1=Yes)')
plt.ylabel('Sex (0=Female, 1=Male)')
plt.title('Average Tip Amount by Sex and Smoker Status')
plt.show()

# Plot 4: Total Bill and Tip Percentage by Weekend or Weekday
pivot4.plot(kind='bar', figsize=(10, 5), color=['skyblue', 'salmon'])
plt.xticks(ticks=[0, 1], labels=['Weekday', 'Weekend'], rotation=0)
plt.xlabel('Is Weekend')
plt.ylabel('Value')
plt.title('Total Bill and Tip Percentage by Weekend or Weekday')
plt.legend(['Total Bill', 'Tip Percentage'])
plt.show()

# Plot 5: Number of Customers Served by Day
plt.figure(figsize=(10, 5))
sns.barplot(x=pivot5.index, y=pivot5['size'], palette='viridis')
plt.xlabel('Day')
plt.ylabel('Number of Customers Served')
plt.title('Number of Customers Served by Day')
plt.show()
```
